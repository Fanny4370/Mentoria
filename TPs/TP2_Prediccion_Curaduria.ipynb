 En esta sección se realizara un análisis de posibles nuevas variables en funcion a la transformación de datos existentes a fin que pueda contribuir al modelo a diseñar, pero antes los invito a realizar un analisis del dato Fecha de Transaccion, ¿esta considerado como un dato fecha? ¿hay alguna otra variable que se le deba cambiar el formato para su tratamiento? Se podria definir nuevas variables centradas en:

(cantidad de transacciones por cliente), que dia de la semana interactua mas, en que hora, cual es el rubro que mas consume a la hora de comprar, donde compra mas, ¿es un cliente que interactua en forma virtual en las compras o prefiere realizar compra presencial? etc.  

b. Siéntanse libres a la hora de definir nuevas variables, pues estas ayudaran al modelo a predecir mejor. Existen muchas herramientas que pueden utilizar para la crecion de nuevas variables como la libreria de Python *featuretools*, los invito a investigar al respecto.

Otra de las técnicas es la utilización del método PCA (análisis de componentes principales) para incluir nuevas variables.

Para la utilización de este método se debe recordar lo siguiente, muchas veces nos enfrentamos con bases de datos cuyas variables se miden en distintas unidades. Por ejemplo para este set de datos tenemos variables que representan importes y otras que tienen valores discretos o incluso valores dicotómicos (0 o 1). Puede que midamos el importe de una transaccion de caja de ahorro de los clientes de la base de datos (valores rondando los 60 mil la media) y a la vez otra variable como es el rubro del comercio (valores rondando 500 a 9400, etc). Como el desvío estándar se mide en las mismas unidades que la variable, una unidad de desvío estándar para el importe de la transaccion no nos dice la misma información que una unidad de desvío estándar para el rubro del comercio.

Esto puede llegar a ser un problema a la hora de utilizar PCA, ya que las escalas de las variables pueden ser muy distintas y por lo tanto pueden gobernar la forma de la nube de datos. Las componentes principales se alinearán entonces a los ejes de las variables con mayor rango de escala, que a lo mejor no son las variables de mayor variabilidad respecto a su valor medio. Es por esto que se suelen estandarizar los datos previo a implementar el PCA.
